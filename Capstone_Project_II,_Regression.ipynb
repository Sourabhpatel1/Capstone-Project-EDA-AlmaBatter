<center><h1><b>Outliers</b></h1></center>
<hr>
<h3><b>From the above boxplots we can see that the following columns have outliers :</b></h3>
<ol>
  <li>Rented Bike Count</li>
  <li>Wind Speed</li>
  <li>Solar Radiation</li>
</ol>

```python
# making a function for removing outliers

def remove_outliers(df):
  for col in df.columns:
    if col not in categorical_data.columns:
      # Calculation the Q1, Q3, IQR, upper and lower range for outliers.
      Q1 = np.quantile(df[col], 0.25)
      Q3 = np.quantile(df[col], 0.75)
      IQR = Q3-Q1
      upper_range = Q3 + 1.5 * IQR
      lower_range = Q1 - 1.5 * IQR
      # Filtering rows based on upper and lower range
      new_df = df.loc[(df[col] >= lower_range) & (df[col] <= upper_range)]
      return new_df
```

```python
data = remove_outliers(data)
```
